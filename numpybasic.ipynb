```python
import numpy as np
a=[[1,2,3,4,4,5],[2,4,2,2,4,5]]
b=np.array(a)
print(b)
print(b.dtype)#type of the array
print(b.ndim)#dimentions
print(b.shape)#(rows,columns)
print(b.size)#number values or row*columns
print(b.T)#row into columns
print(b.itemsize)#if it dtype=64 then it is 8 ,if it is 32 then it is 4
print(b.nbytes)#nbytes=size*itemsize
```
